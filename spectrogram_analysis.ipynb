```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy import signal
from matplotlib.colors import Normalize, LogNorm

import auxlib
```

```python
saving_figs = os.path.join(os.getcwd(), 'figures/spectrogram_analysis')
if not os.path.exists(saving_figs):
	os.makedirs(saving_figs)

saving_files = os.path.join(os.getcwd(), 'media/spectrogram_analysis')
if not os.path.exists(saving_files):
	os.makedirs(saving_files)
```

```python
channels = [10, 31, 36, 42, 43, 48, 53, 63]
```

```python
sujeto = 1
run = 3

sujeto_name = 'S' + str(sujeto).zfill(3) + 'R' + str(run).zfill(2)

data, signal_headers, header = auxlib.loadEEG(subject=sujeto, record=run)
```

```python
fs = 160.0
fig, axs = plt.subplots(2, 4, figsize=(16, 8), sharex=True, sharey=True)
norm = Normalize(vmin=0, vmax=10000)

# Loop through channels 1 to 64
for ch in channels:
	x = data[ch]
	indice = channels.index(ch)
	f, t, Sxx = signal.spectrogram(x, fs, nfft=1024, noverlap=32)

	axs[(indice)//4, (indice)%4].pcolormesh(t, f, Sxx, shading='gouraud', cmap='viridis', norm=norm)
	axs[(indice)//4, (indice)%4].set_ylabel('Frequency [Hz]')
	axs[(indice)//4, (indice)%4].set_ylim(0, 7)
	axs[(indice)//4, (indice)%4].set_xlabel('Time [sec]')
	axs[(indice)//4, (indice)%4].set_title("Ch {}".format(ch+1))

# Layout so plots do not overlap
fig.tight_layout()
fig.savefig(os.path.join(saving_figs, f'{sujeto_name}_spectrogram_channels.png'), dpi=300, bbox_inches='tight')
# Show the plot
plt.show()

pd.DataFrame(data).to_csv(os.path.join(saving_files, f'{sujeto_name}_spectrogram.csv'))
```
